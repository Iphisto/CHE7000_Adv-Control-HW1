```python
#QUESTION 2

# (a)Finding the max, min, and sum
list_a = [3, 4, 6, 10, 39, 87, 12, 54]

max_list_a = max(list_a)
min_list_a= min(list_a)
sum_list_a = sum(list_a)

print(f"Maximum value: {max_list_a}")
print(f"Minimum value: {min_list_a}")
print(f"Sum: {sum_list_a}")

# (b)sorting
list_a.sort()
print(f"Sorted list values: {list_a}")

# (c)loop
print("Elements in list_a:")
for i in list_a:
    print(i, end=' ')
```

```python
#QUESTION 3

# a) prime number function
def is_prime(num):
    if num <= 1:
        return False
    # Check factors from 2 up to the square root of the number
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# b) prime numbers below 100; let p be the list of prime numbers below 100
p = [x for x in range(100) if is_prime(x)]
print(f"Primes below 100: {p}")

# c) prime numbers below n
def primenum_below_n(n):
    primenum = []
    for x in range(n):
        if is_prime(x):
            primenum.append(x)
    return primenum

# d) when n = 1000
primenum_below_1000 = primenum_below_n(1000)

print(f"Number of primes below 1000: {len(primenum_below_1000)}")
print(f"prime numbers below 1000: {primenum_below_1000}")
```

```python
#QUESTION 4

import numpy as np

# a) 3D array
def a_3d_array(n, data):

    if data == 'zeros' or data == 0:
        return np.zeros((n, 3, 3))
    else:
      a,b = data
      return np.random.randint(a, b, size=(n, 3, 3))

# b)
array_zeros = a_3d_array(10, 0)
print(f"n= 10 and 3D array contains zeros: {array_zeros.shape}")
print(array_zeros)

# c)
array_random = a_3d_array(5, data=[1,100])
print("n = 5, random numbers between 1 and 100", array_random)
```

```python
#QUESTION 5
import numpy as np

# USING A\B
A = np.array([
    [3, 4, -5],
    [0, 5, 1],
    [1, 3, 6]
])

B = np.array([10, 8, -5])

# Solve for X [a, b, c]
X = np.linalg.solve(A, B)

a, b, c = X
print(f"Solution:\na = {a:.2f}\nb = {b:.2f}\nc = {c:.2f}")
```

```python
#QUESTION 6

import matplotlib.pyplot as plt
import numpy as np


def fun(x):
    return 0.1 * x**3 - x**2 + 5

x = np.linspace(-5, 5, 100)
y = fun(x)

# b) max and min values
y_max = np.max(y)
x_max = x[np.argmax(y)]

y_min = np.min(y)
x_min = x[np.argmin(y)]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$y = 0.1x^3 - x^2 + 5$', color='blue')

# points for max and min
plt.scatter([x_max], [y_max], color='green', zorder=5)
plt.scatter([x_min], [y_min], color='red', zorder=5)

# Annotating the points
plt.text(x_max, y_max + 1, f'Max\n({x_max:.1f}, {y_max:.1f})', ha='center', color='green')
plt.text(x_min, y_min - 3, f'Min\n({x_min:.1f}, {y_min:.1f})', ha='center', color='red')

plt.title("Plot of y = 0.1x³ - x² + 5")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print(f"Maximum value in range: {y_max:.2f} at x = {x_max:.2f}")
print(f"Minimum value in range: {y_min:.2f} at x = {x_min:.2f}")
```
